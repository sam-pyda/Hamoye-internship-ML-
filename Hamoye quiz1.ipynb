```python
#Q13: What is the total number and percentage of missing data in 2016 to 2 decimal places?
x= df.isnull().sum()
missing_2016 = x['Y2016']
total_missing =df.isnull().sum().sum()
percentage_missing_2016 = missing_2016 / total_missing *100
print(f'the total number and percentage of missing data in 2016 is {missing_2016} and {percentage_missing_2016}% respectively')
```

```python
#Q14: Which year had the highest correlation with ‘Element Code’? 
a= df[['Element Code', 'Y2018','Y2015','Y2014','Y2016','Y2017']]

a.corr()
```

```python
#Q15: What year has the highest sum of Import Quantity?
#Hint- Perform a groupby operation on ‘Element’ and use the resulting Dataframe to answer the question
df.groupby('Element').sum()
```

```python
# Q17: Which of these elements had the highest sum in 2018?
df1= df[['Element','Y2018']]
Max_element_2018= df1.groupby('Element').sum().idxmax()
print(Max_element_2018)
```

```python
# Q18: Which of these elements had the 3rd lowest sum in 2018?
third_lowest_element_2018 = df1.groupby('Element').sum().sort_values('Y2018',ascending=True)
third_lowest_element_2018
```

```python
print('The elements having the 3rd lowest sum in 2018 is : Protein supply quantity (g/capita/day)')
```
